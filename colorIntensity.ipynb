```python
import PIL
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageDraw
from PIL import ImageFont


image=Image.open("bob.png")
image=image.convert('RGB')
px = image.load()
width, height = image.size

newImage = PIL.Image.new(image.mode, (width, (height + 50)))
newImage.paste(image, (0, 0))



images=[]
image2 = newImage.copy()
image3 = newImage.copy()
image4 = newImage.copy()
for x in range(height):
    for i in range(width):
        r, g, b = newImage.getpixel((i,x))
        image2.putpixel( (i, x), (int(r*0.1), g, b))
        image3.putpixel( (i, x), (int(r*0.5), g, b))
        image4.putpixel( (i, x), (int(r*0.9), g, b))     
images.append(image2)
images.append(image3)
images.append(image4)

image5 = newImage.copy()
image6 = newImage.copy()
image7 = newImage.copy()
for x in range(height):
    for i in range(width):
        r, g, b = newImage.getpixel((i,x))
        image5.putpixel( (i, x), (r, int(g*0.1), b))
        image6.putpixel( (i, x), (r, int(g*0.5), b))
        image7.putpixel( (i, x), (r, int(g*0.9), b))     
images.append(image5)
images.append(image6)
images.append(image7)

image8 = newImage.copy()
image9 = newImage.copy()
image10 = newImage.copy()
for x in range(height):
    for i in range(width):
        r, g, b = newImage.getpixel((i,x))
        image8.putpixel( (i, x), (r, g, int(b*0.1)))
        image9.putpixel( (i, x), (r, g, int(b*0.5)))
        image10.putpixel( (i, x), (r, g, int(b*0.9)))     
images.append(image8)
images.append(image9)
images.append(image10)


intensity  = [0.1, 0.5, 0.9]
i = 0
for x in images[:3]:
        draw = ImageDraw.Draw(x)
        font = ImageFont.truetype("fanwood-webfont.ttf", 50)
        text = 'channel 0 intensity {}'.format(intensity[i])
        draw.text((5, 560), text, fill = (int(188*intensity[i]), 206, 212), font = font, align ="left")
        i = i+1
i = 0        
for x in images[3:6]:
        draw = ImageDraw.Draw(x)
        font = ImageFont.truetype("fanwood-webfont.ttf", 50)
        text = 'channel 1 intensity {}'.format(intensity[i])
        draw.text((5, 560), text, fill = (188, int(206*intensity[i]), 212), font = font, align ="left")
        i = i+1
i = 0        
for x in images[6:]:
        draw = ImageDraw.Draw(x)
        font = ImageFont.truetype("fanwood-webfont.ttf", 50)
        text = 'channel 2 intensity {}'.format(intensity[i])
        draw.text((5, 560), text, fill = (188, 206, int(212*intensity[i])), font = font, align ="left")
        i = i+1


first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
x=0
y=0

for img in images:

    contact_sheet.paste(img, (x, y) )
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width

contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
